```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/householdtask3.csv" )

display (data.head(10))
```

```python

import matplotlib. pyplot as plt
```

```python
data.columns
```

```python
#bar chart
plt.figure(figsize=(10, 6))
plt.bar(data['year'], data['income'], label='Total Income')
plt.xlabel('Year')
plt.ylabel('Total Income')
plt.title('Total Income by Year')
plt.legend()
plt.show()
```

```python
#line chart
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['own'], marker='o', label='Own Households')
plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Own Households by Year')
plt.legend()
plt.grid(True)
plt.show()
```
